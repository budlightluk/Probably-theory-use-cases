```python
import random

def coin_toss():
    return random.choice(["Орел", "Решка"])

trials = 10000
results = [coin_toss() for _ in range(trials)]
orel_count = results.count("Орел")
reshka_count = results.count("Решка")

print(f"Вероятность выпадения орла: {orel_count / trials:.2f}")
print(f"Вероятность выпадения решки: {reshka_count / trials:.2f}")

```

```python
import random

def roll_dice():
    return random.randint(1, 6)

trials = 10000
results = [roll_dice() for _ in range(trials)]
counts = [results.count(i) for i in range(1, 7)]

for i, count in enumerate(counts):
    print(f"Вероятность выпадения {i+1}: {count / trials:.2f}")
```

Вероятность выпадения рояля в покере

```python
import numpy as np

def markov_chain(transition_matrix, initial_state):
    state = initial_state
    for _ in range(100):
        state = np.random.choice(len(transition_matrix), p=transition_matrix[state])
    return state

transition_matrix = [[0.7, 0.3], [0.4, 0.6]]
initial_state = 0
final_state = markov_chain(transition_matrix, initial_state)
print(f"Вероятность перехода в состояние {final_state}: {transition_matrix[initial_state][final_state]:.2f}")
```

Симуляция марковской цепи

```python
import fractions

def ball_probability(color):
    if color == "красный":
        return fractions.Fraction(5, 10)
    elif color == "синий":
        return fractions.Fraction(3, 10)
    elif color == "зеленый":
        return fractions.Fraction(2, 10)

color = "красный"
print(f"Вероятность выпадения {color} шара: {ball_probability(color)}")
```

Вероятность выпадения определенного цвета шара
Задача: Вычислить вероятность выпадения определенного цвета шара из урны,
содержащей 5 красных, 3 синих и 2 зеленых шара.

```python
import numpy as np
import matplotlib.pyplot as plt
pip install --upgrade matplotlib

def normal_distribution(mean, std_dev):
    return np.random.normal(mean, std_dev)

mean = 0
std_dev = 1
trials = 10000
results = [normal_distribution(mean, std_dev) for _ in range(trials)]
counts = np.histogram(results, bins=100, range=(-3, 3), density=True)[0]

cdf = np.cumsum(counts) / np.sum(counts)

plt.plot(np.linspace(-3, 3, 100), cdf)
plt.xlabel('Значение')
plt.ylabel('Кумулятивная вероятность')
plt.title('Кумулятивная функция распределения')
plt.show()
```

```python
import fractions

def two_dice_probability(target_value):
    count = 0
    for i in range(1, 7):
        for j in range(1, 7):
            if i + j == target_value:
                count += 1
    return fractions.Fraction(count, 36)

target_value = 7
print(f"Вероятность выпадения значения {target_value}: {two_dice_probability(target_value)}")
```

```python
import numpy as np

def poisson_distribution(lambda_):
    return np.random.poisson(lambda_)

lambda_ = 5
trials = 10000
results = [poisson_distribution(lambda_) for _ in range(trials)]
counts = np.histogram(results, bins=10, range=(0, 10))[0]

for i, count in enumerate(counts):
    print(f"Вероятность выпадения {i} событий: {count / trials:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def poisson_distribution(lambda_):
    return np.random.poisson(lambda_)

lambda_ = 5
trials = 10000
results = [poisson_distribution(lambda_) for _ in range(trials)]
counts = np.histogram(results, bins=10, range=(0, 10))[0]

plt.bar(np.arange(10), counts / trials, width=1)
plt.xlabel('Количество событий')
plt.ylabel('Вероятность')
plt.title('Распределение Пуассона')
plt.show()
```

Распределение Пуассона с гистограммой

```python
import math

def bernoulli_process_probability(success_probability, trials, target_successes):
    return math.comb(trials, target_successes) * (success_probability ** target_successes) * ((1 - success_probability) ** (trials - target_successes))

success_probability = 0.7
trials = 10
target_successes = 7
print(f"Вероятность выпадения {target_successes} успехов: {bernoulli_process_probability(success_probability, trials, target_successes):.2f}")
```
